```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

file_path = '/content/cirrhosis.csv'
df = pd.read_csv(file_path)
```

```python
df.replace('NA', np.nan, inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df.fillna(df.select_dtypes(include=[np.number]).mean(), inplace=True)
```

```python

df.fillna(df.select_dtypes(include=[np.number]).mean(), inplace=True)
df.fillna(df.select_dtypes(include=['category','object']).mode().iloc[0], inplace=True)
```

```python
category_colums=['Status', 'Drug', 'Sex', 'Ascites', 'Hepatomegaly', 'Spiders', 'Edema', 'Stage']
for col in category_colums :
  df[col] = df[col].astype('category')
```

```python
df['Age_years'] = df['Age']/365
```

```python
#Exploratory Data Analysis (EDA)
```

```python
print(df.describe())
```

```python
print(df.describe(include='category'))
```

```python
plt.figure(figsize=(10,6))
```

```python
sns.countplot(x='Status', data=df)
plt.title('Count of Patient Status')
plt.show()
```

```python
sns.pairplot(df[['Bilirubin', 'Albumin', 'Copper', 'Cholesterol', 'Status']], hue='Status')
plt.show()
```

```python
numerical_cols=['N_Days', 'Bilirubin', 'Cholesterol', 'Albumin', 'Copper', 'Alk_Phos', 'SGOT', 'Tryglicerides', 'Platelets', 'Prothrombin']
df[numerical_cols].hist(bins=15, figsize=(15,10), layout=(5,3))
plt.suptitle('Distribution of Numerical Features')
plt.show()
#This is distribution of numerical features
```

```python
#boxplot
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Status', y=col, data=df)
    plt.title(f'Box Plot of {col}')
    plt.show()
```

```python
#pairplot
sns.pairplot(df[['Bilirubin', 'Albumin', 'Copper', 'Cholesterol', 'Status']], hue='Status')
plt.show()
```

```python
# Correlation matrix
corr_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Violin Plots for Numerical Features
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.violinplot(x='Status', y=col, data=df)
    plt.title(f'Violin Plot of {col}')
    plt.show()
```

```python
status_drug_counts = df.groupby(['Status', 'Drug']).size().unstack()
status_drug_counts.plot(kind='bar', stacked=True, figsize=(10,6))
plt.xlabel('Status')
plt.ylabel('Count')
plt.title('Drug Distribution by Status')
plt.show
# Stacked bar plot for Status and Drug
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.show()
```
